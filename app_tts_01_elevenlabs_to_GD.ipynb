```python
import requests
import time

# Opcional: si tienes implementado un módulo de errores en la librería
try:
    from elevenlabs.errors import ElevenLabsAPIError
except ImportError:
    ElevenLabsAPIError = None

# Opcional: función para cargar la API key del entorno
def load_api_key() -> str:
    """
    Carga la API key desde una variable de entorno o desde un archivo de configuración.
    Este es un ejemplo básico; se puede adaptar según las necesidades.
    """
    import os
    return os.getenv("ELEVENLABS_API_KEY", "")

def elevenlabs_generate_audio_str(config: dict) -> str:
    """
    Genera un archivo de audio a partir de un texto usando la API de ElevenLabs y guarda el archivo en la ruta especificada.

    Args:
        config (dict):
            - text_str (str): Texto a convertir en audio.
            - api_key_str (str): Clave API para autenticarse. Si no se provee, se intenta cargar desde el entorno.
            - voice_id_str (str, opcional): Identificador de la voz (default: 'default_voice_id').
            - drive_path_str (str, opcional): Ruta para guardar el archivo de audio
                                               (default: '/content/drive/MyDrive/audio_output.mp3').
            - retry_attempts (int, opcional): Número de reintentos en caso de error (default: 3).
            - retry_delay (int, opcional): Segundos a esperar entre reintentos (default: 2).

    Returns:
        str: Mensaje indicando la ruta del archivo de audio generado.

    Raises:
        ValueError: Si faltan parámetros obligatorios o si se superan los reintentos.
        ElevenLabsAPIError: Si se dispone de este error personalizado y se alcanza la cuota.
    """
    # Validación de parámetros
    text_str = config.get('text_str')
    if not text_str:
        raise ValueError("[VALIDATION [ERROR ❌]] Falta 'text_str' en config.")

    api_key_str = config.get('api_key_str')
    if not api_key_str:
        api_key_str = load_api_key()
        if not api_key_str:
            raise ValueError("[VALIDATION [ERROR ❌]] Falta 'api_key_str' en config y no se pudo cargar desde el entorno.")

    voice_id_str = config.get('voice_id_str', 'default_voice_id')
    drive_path_str = config.get('drive_path_str', '/content/drive/MyDrive/audio_output.mp3')
    retry_attempts = config.get('retry_attempts', 3)
    retry_delay = config.get('retry_delay', 2)

    print("🔹🔹🔹 [START ▶️] Generación de audio con ElevenLabs 🔹🔹🔹", flush=True)

    # Construcción de la URL y configuración de la petición
    url_str = f"https://api.elevenlabs.io/v1/text-to-speech/{voice_id_str}"
    headers_dic = {
        "xi-api-key": api_key_str,
        "Content-Type": "application/json"
    }
    data_dic = {
        "text": text_str
        # Se pueden incluir parámetros adicionales según la documentación
    }

    # Realizar la llamada a la API con reintentos
    attempt = 0
    response = None
    while attempt < retry_attempts:
        try:
            response = requests.post(url_str, json=data_dic, headers=headers_dic)
            if response.status_code == 200:
                break
            else:
                error_msg = (f"[API ERROR ❌] Error en la llamada a la API: "
                             f"{response.status_code} {response.text}")
                print(error_msg, flush=True)
                attempt += 1
                if attempt < retry_attempts:
                    print(f"[RETRY 🔄] Reintentando en {retry_delay} segundos... "
                          f"(Intento {attempt+1}/{retry_attempts})", flush=True)
                    time.sleep(retry_delay)
                else:
                    if ElevenLabsAPIError:
                        raise ElevenLabsAPIError(error_msg)
                    else:
                        raise ValueError(error_msg)
        except Exception as e:
            attempt += 1
            print(f"[API CALL ERROR ❌] Excepción durante la llamada a la API: {e}", flush=True)
            if attempt < retry_attempts:
                print(f"[RETRY 🔄] Reintentando en {retry_delay} segundos... "
                      f"(Intento {attempt+1}/{retry_attempts})", flush=True)
                time.sleep(retry_delay)
            else:
                raise ValueError(f"[API CALL ERROR ❌] No se pudo generar el audio después de "
                                 f"{retry_attempts} intentos: {e}")

    # Guardar el archivo de audio en la ruta especificada
    try:
        with open(drive_path_str, 'wb') as audio_file:
            audio_file.write(response.content)
        print(f"[FILE SAVE SUCCESS ✅] Archivo de audio guardado en: {drive_path_str}", flush=True)
    except Exception as e:
        raise ValueError(f"[FILE SAVE ERROR ❌] No se pudo guardar el archivo: {e}")

    print("🔹🔹🔹 [FINISHED ✅] Generación de audio completada 🔹🔹🔹", flush=True)
    return f"Audio generado y guardado en: {drive_path_str}"

```

```python
import os
import requests
import pandas as pd
from google.cloud import secretmanager

def elevenlabs_voice_list(config: dict) -> pd.DataFrame:
    """
    Recupera la lista de voces disponibles en ElevenLabs haciendo una llamada directa
    al endpoint GET /v1/voices utilizando el parámetro 'limit' definido en config, y la transforma en un DataFrame.

    Utiliza la autenticación vía GCP Secret Manager para obtener la API key de ElevenLabs. Se espera que en
    config estén definidas las siguientes claves:
      - "ini_environment_identificated": Identificador del entorno ("LOCAL", "COLAB", etc.).
      - "json_keyfile_local": Ruta al archivo de credenciales para entorno LOCAL.
      - "json_keyfile_colab": Ruta al archivo de credenciales para entorno COLAB.
      - "json_keyfile_GCP_secret_id": Ruta o ID para GCP Secret Manager.
      - "TTS_API_key_GCP_secret_manager_name": Nombre del secreto (ej: "elevenlabs").
      - "TTS_API_key_GCP_secret_manager_project_id": ID del proyecto donde se encuentra el secreto (ej: "animum-dev-apps").
      - "limit": (opcional) Número máximo de voces a solicitar.

    Returns:
        pd.DataFrame: DataFrame con la información de las voces.

    Raises:
        ValueError: Si falta algún parámetro o si ocurre un error durante la autenticación o la obtención de voces.
    """
    try:
        # Validar parámetros obligatorios
        required_keys = [
            "ini_environment_identificated",
            "json_keyfile_local",
            "json_keyfile_colab",
            "json_keyfile_GCP_secret_id",
            "TTS_API_key_GCP_secret_manager_name",
            "TTS_API_key_GCP_secret_manager_project_id"
        ]
        for key in required_keys:
            if key not in config:
                raise ValueError(f"[VALIDATION ERROR] Falta '{key}' en config.")

        # Obtener el parámetro limit (por defecto 100)
        limit = config.get("limit", 100)

        def _retrieve_tts_api_key_from_secret_manager(config: dict) -> str:
            """
            Recupera la API key para ElevenLabs (TTS) desde GCP Secret Manager.

            Utiliza:
              - TTS_API_key_GCP_secret_manager_name: Nombre del secreto (ej: "elevenlabs")
              - TTS_API_key_GCP_secret_manager_project_id: ID del proyecto (ej: "animum-dev-apps")

            Returns:
                str: La API key recuperada.

            Raises:
                ValueError: Si ocurre un error durante la recuperación.
            """
            project_id = config.get("TTS_API_key_GCP_secret_manager_project_id")
            secret_name = config.get("TTS_API_key_GCP_secret_manager_name")
            if not project_id or not secret_name:
                raise ValueError("[VALIDATION ERROR] Faltan parámetros para Secret Manager.")

            ini_env = config.get("ini_environment_identificated")
            if ini_env == "LOCAL":
                os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = config.get("json_keyfile_local")
            elif ini_env == "COLAB":
                os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = config.get("json_keyfile_colab")
            else:
                os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = config.get("json_keyfile_GCP_secret_id")

            secret_full_name = f"projects/{project_id}/secrets/{secret_name}/versions/latest"
            print(f"[SECRET MANAGER] Accediendo al secreto de TTS: {secret_full_name}", flush=True)
            try:
                client_sm = secretmanager.SecretManagerServiceClient()
                response = client_sm.access_secret_version(request={"name": secret_full_name})
                api_key = response.payload.data.decode("UTF-8")
                print("[SECRET MANAGER] TTS_API_key obtenida correctamente.", flush=True)
                return api_key
            except Exception as e:
                raise ValueError(f"[SECRET MANAGER ERROR] Error al obtener TTS_API_key: {e}")

        # Recuperar la API key de ElevenLabs
        tts_api_key = _retrieve_tts_api_key_from_secret_manager(config)
        if not tts_api_key:
            raise ValueError("[VALIDATION ERROR] No se pudo recuperar la TTS API key.")

        # Realizar la solicitud GET al endpoint, utilizando el parámetro limit
        url = f"https://api.elevenlabs.io/v1/voices?limit={limit}"
        headers = {"xi-api-key": tts_api_key}
        response = requests.get(url, headers=headers)
        if response.status_code != 200:
            raise ValueError(f"[API ERROR] Código: {response.status_code}. Respuesta: {response.text}")

        data = response.json()
        voices = data.get("voices", [])

        # Convertir la lista de voces a DataFrame utilizando pd.json_normalize
        df = pd.json_normalize(voices)
        return df

    except Exception as e:
        raise ValueError(f"[ERROR] {e}")
```
