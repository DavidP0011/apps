```python
def _ini_authenticate_API(config: dict, project_id: str):
    """
    Autentica utilizando el diccionario com√∫n en config.

    Dependiendo de 'ini_environment_identificated' se utiliza:
      - "LOCAL": usa la key "json_keyfile_local"
      - "COLAB": usa la key "json_keyfile_colab"
      - Para GCP (por ejemplo, "COLAB_ENTERPRISE" o un project_id): usa "json_keyfile_GCP_secret_id"

    Args:
      config (dict): Diccionario de configuraci√≥n.
      project_id (str): ID del proyecto GCP (se usa en la autenticaci√≥n GCP).

    Returns:
      Credentials: Objeto de credenciales para la autenticaci√≥n.
    """
    from google.oauth2 import service_account

    env = config.get("ini_environment_identificated", "COLAB")
    if env == "LOCAL":
        json_path = config.get("json_keyfile_local")
        if not json_path:
            raise ValueError("[AUTHENTICATION ERROR ‚ùå] Falta 'json_keyfile_local' en config para entorno LOCAL.")
        credentials = service_account.Credentials.from_service_account_file(json_path)
    elif env == "COLAB":
        json_path = config.get("json_keyfile_colab")
        if not json_path:
            raise ValueError("[AUTHENTICATION ERROR ‚ùå] Falta 'json_keyfile_colab' en config para entorno COLAB.")
        credentials = service_account.Credentials.from_service_account_file(json_path)
    else:
        # Asumimos que para GCP (COLAB_ENTERPRISE o si se pasa un project_id distinto) se usa Secret Manager.
        secret_id = config.get("json_keyfile_GCP_secret_id")
        if not secret_id:
            raise ValueError("[AUTHENTICATION ERROR ‚ùå] Falta 'json_keyfile_GCP_secret_id' en config para entornos GCP.")
        from google.cloud import secretmanager
        client_sm = secretmanager.SecretManagerServiceClient()
        secret_name = f"projects/{project_id}/secrets/{secret_id}/versions/latest"
        response = client_sm.access_secret_version(name=secret_name)
        secret_str = response.payload.data.decode("UTF-8")
        import json
        secret_info = json.loads(secret_str)
        credentials = service_account.Credentials.from_service_account_info(secret_info)
    return credentials
```

```python
def TTS_sintetizar_texto_str(config: dict) -> str:
    """
    Sintetiza el texto proporcionado a voz utilizando la API de Google Cloud Text-to-Speech y guarda el audio en archivos MP3.
    Permite elegir entre entrada de texto regular o SSML, seg√∫n la clave 'text_input_type'.
    Para cada voz en 'voice_name_list', se genera un archivo cuyo nombre es el mismo que el de la voz.
    Se verifica que la carpeta de salida exista y se tenga permiso de escritura antes de llamar a la API.

    Args:
        config (dict):
            - text_regular (str): Texto a sintetizar en modo regular.
            - text_SSML (str): Texto en formato SSML a sintetizar.
            - text_input_type (str): Tipo de entrada, "regular" o "ssml". Default: "regular".
            - salida (str, opcional): Ruta de la carpeta de salida en Google Drive. Ejemplo: "/content/drive/MyDrive/tu_carpeta".
              Si la carpeta no existe, se crear√°. (Default: "output")
            - language_code (str, opcional): C√≥digo de idioma para la voz (default: "es-ES").
            - ssml_gender (str, opcional): G√©nero de la voz cuando se usa texto regular. Opciones: "NEUTRAL", "MALE", "FEMALE" (default: "NEUTRAL").
            - voice_name_list (list, opcional): Lista de nombres espec√≠ficos de voces (consulte
              https://cloud.google.com/text-to-speech/docs/voices?hl=es-419). Se genera un archivo MP3 para cada voz.
              Si se deja vac√≠a, se utiliza la configuraci√≥n por defecto.

    Returns:
        str: Mensaje indicando la ubicaci√≥n de los archivos de audio generados.

    Raises:
        ValueError: Si faltan par√°metros obligatorios o si 'voice_name_list' no es una lista.
        Exception: Si ocurre un error durante la verificaci√≥n de escritura o la s√≠ntesis de voz.
    """
    import os

    # Subfunci√≥n: Validaci√≥n y extracci√≥n de par√°metros
    def _validate_config(config: dict) -> tuple:
        text_input_type = config.get("text_input_type", "regular").lower()
        if text_input_type not in ("regular", "ssml"):
            raise ValueError("[VALIDATION [ERROR ‚ùå]] 'text_input_type' debe ser 'regular' o 'ssml'.")
        if text_input_type == "ssml":
            texto_str = config.get("text_SSML")
            if not texto_str:
                raise ValueError("[VALIDATION [ERROR ‚ùå]] Falta 'text_SSML' en config cuando text_input_type es 'ssml'.")
        else:
            texto_str = config.get("text_regular")
            if not texto_str:
                raise ValueError("[VALIDATION [ERROR ‚ùå]] Falta 'text_regular' en config cuando text_input_type es 'regular'.")
        salida_str = config.get("salida", "output")
        language_code_str = config.get("language_code", "es-ES")
        ssml_gender_str = config.get("ssml_gender", "NEUTRAL").upper()
        voice_name_list = config.get("voice_name_list", [])
        if not isinstance(voice_name_list, list):
            raise ValueError("[VALIDATION [ERROR ‚ùå]] 'voice_name_list' debe ser una lista.")
        return texto_str, text_input_type, salida_str, language_code_str, ssml_gender_str, voice_name_list

    # Subfunci√≥n: Mapeo de g√©nero de voz a la enumeraci√≥n de la API
    def _map_ssml_gender(ssml_gender_str: str):
        from google.cloud import texttospeech
        gender_mapping = {
            "NEUTRAL": texttospeech.SsmlVoiceGender.NEUTRAL,
            "MALE": texttospeech.SsmlVoiceGender.MALE,
            "FEMALE": texttospeech.SsmlVoiceGender.FEMALE,
        }
        return gender_mapping.get(ssml_gender_str, texttospeech.SsmlVoiceGender.NEUTRAL)

    # Subfunci√≥n: Verifica el permiso de escritura en la carpeta de salida
    def _check_write_access(folder_path: str) -> None:
        test_file = os.path.join(folder_path, "temp_write_test.txt")
        try:
            with open(test_file, "w") as f:
                f.write("test")
            os.remove(test_file)
            print(f"[FILE INFO ‚ÑπÔ∏è] Permiso de escritura confirmado en: '{folder_path}'", flush=True)
        except Exception as e:
            raise Exception(f"[FILE ERROR ‚ùå] No se puede escribir en la carpeta '{folder_path}': {e}")

    # Subfunci√≥n: Realiza la s√≠ntesis de voz para un nombre de voz espec√≠fico
    def _synthesize_speech(texto_str: str, language_code_str: str, ssml_gender_enum, voice_name_str: str, text_input_type: str) -> any:
        from google.cloud import texttospeech
        # Seleccionar la entrada de texto seg√∫n el tipo: regular o SSML
        if text_input_type == "ssml":
            input_text = texttospeech.SynthesisInput(ssml=texto_str)
        else:
            input_text = texttospeech.SynthesisInput(text=texto_str)
        # Si se especifica voice_name_str, se utiliza; de lo contrario, se usa la configuraci√≥n por defecto
        if voice_name_str:
            voice = texttospeech.VoiceSelectionParams(
                language_code=language_code_str,
                name=voice_name_str
            )
        else:
            voice = texttospeech.VoiceSelectionParams(
                language_code=language_code_str,
                ssml_gender=ssml_gender_enum
            )
        audio_config = texttospeech.AudioConfig(
            audio_encoding=texttospeech.AudioEncoding.MP3
        )
        print(f"[TTS INFO ‚ÑπÔ∏è] Solicitando s√≠ntesis para voz: '{voice_name_str or 'default'}'", flush=True)
        client = texttospeech.TextToSpeechClient()
        return client.synthesize_speech(
            input=input_text,
            voice=voice,
            audio_config=audio_config
        )

    # Subfunci√≥n: Guarda el contenido de audio en un archivo
    def _save_audio_file(audio_content: bytes, file_path: str) -> None:
        with open(file_path, "wb") as out:
            out.write(audio_content)
        print(f"[TTS SUCCESS ‚úÖ] Audio guardado en: '{file_path}'", flush=True)

    try:
        texto_str, text_input_type, salida_folder_str, language_code_str, ssml_gender_str, voice_name_list = _validate_config(config)
        ssml_gender_enum = _map_ssml_gender(ssml_gender_str)

        # Asegurar que la carpeta de salida exista y sea un directorio
        if not os.path.exists(salida_folder_str):
            os.makedirs(salida_folder_str, exist_ok=True)
            print(f"[FILE INFO ‚ÑπÔ∏è] Carpeta creada: '{salida_folder_str}'", flush=True)
        elif not os.path.isdir(salida_folder_str):
            raise Exception(f"[FILE ERROR ‚ùå] La ruta '{salida_folder_str}' no es una carpeta.")

        # Verificar permiso de escritura en la carpeta de salida
        _check_write_access(salida_folder_str)

        print("üîπüîπüîπ [START ‚ñ∂Ô∏è] Iniciando s√≠ntesis de voz para cada voz en 'voice_name_list' üîπüîπüîπ", flush=True)
        output_files_list = []
        if not voice_name_list:
            voice_name_list = [""]

        for voice_name in voice_name_list:
            try:
                response = _synthesize_speech(texto_str, language_code_str, ssml_gender_enum, voice_name, text_input_type)
                file_name = f"{voice_name}.mp3" if voice_name else "output.mp3"
                file_path = os.path.join(salida_folder_str, file_name)
                _save_audio_file(response.audio_content, file_path)
                output_files_list.append(file_path)
            except Exception as e_voice:
                print(f"[TTS ERROR ‚ùå] Error al sintetizar con voz '{voice_name}': {e_voice}", flush=True)
                continue

        if not output_files_list:
            raise Exception("No se gener√≥ ning√∫n archivo de audio.")

        files_str = "\n".join(output_files_list)
        return f"Audio generado correctamente en los siguientes archivos:\n{files_str}"

    except Exception as e:
        print(f"[TTS ERROR ‚ùå] Error durante la s√≠ntesis de voz: {e}", flush=True)
        raise e

# Ejemplo de plantilla de configuraci√≥n para probar SSML (usando el ejemplo de los apuntes)
config_template = {
    "text_input_type": "ssml",  # Usar "ssml" para entrada SSML o "regular" para texto simple
    "text_SSML": (
        "<speak>\n"
        "  <p>\n"
        "    <s>Bienvenidos a la demostraci√≥n de SSML.</s>\n"
        "    <s>Hoy aprenderemos a controlar la entonaci√≥n y el ritmo del discurso.</s>\n"
        "  </p>\n"
        "  <p>\n"
        "    <s>\n"
        "      <prosody rate=\"medium\" pitch=\"+1st\" volume=\"medium\">\n"
        "        Este p√°rrafo se pronuncia con una entonaci√≥n ligeramente elevada.\n"
        "      </prosody>\n"
        "    </s>\n"
        "  </p>\n"
        "  <break time=\"700ms\"/>\n"
        "  <p>\n"
        "    <s>\n"
        "      <emphasis level=\"strong\">¬°Atenci√≥n!</emphasis> Presta mucha atenci√≥n a este ejemplo.\n"
        "    </s>\n"
        "  </p>\n"
        "  <p>\n"
        "    <s>\n"
        "      La fecha de hoy es <say-as interpret-as=\"date\" format=\"dm\">25/12</say-as>.\n"
        "    </s>\n"
        "  </p>\n"
        "</speak>"
    ),
    "text_regular": "Este es un ejemplo de texto regular para s√≠ntesis de voz.",
    "salida": "/content/drive/MyDrive/test_tts3",  # Carpeta de salida en Google Drive
    "language_code": "es-ES",
    "ssml_gender": "NEUTRAL",
    "voice_name_list": voice_name_list
}

# Llamada a la funci√≥n (se recomienda probar primero con pocas voces para validar la configuraci√≥n)
resultado = TTS_sintetizar_texto_str(config_template)
print(resultado)
```
