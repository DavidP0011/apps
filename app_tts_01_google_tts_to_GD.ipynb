```python
def _ini_authenticate_API(config: dict, project_id: str):
    """
    Autentica utilizando el diccionario común en config.

    Dependiendo de 'ini_environment_identificated' se utiliza:
      - "LOCAL": usa la key "json_keyfile_local"
      - "COLAB": usa la key "json_keyfile_colab"
      - Para GCP (por ejemplo, "COLAB_ENTERPRISE" o un project_id): usa "json_keyfile_GCP_secret_id"

    Args:
      config (dict): Diccionario de configuración.
      project_id (str): ID del proyecto GCP (se usa en la autenticación GCP).

    Returns:
      Credentials: Objeto de credenciales para la autenticación.
    """
    from google.oauth2 import service_account

    env = config.get("ini_environment_identificated", "COLAB")
    if env == "LOCAL":
        json_path = config.get("json_keyfile_local")
        if not json_path:
            raise ValueError("[AUTHENTICATION ERROR ❌] Falta 'json_keyfile_local' en config para entorno LOCAL.")
        credentials = service_account.Credentials.from_service_account_file(json_path)
    elif env == "COLAB":
        json_path = config.get("json_keyfile_colab")
        if not json_path:
            raise ValueError("[AUTHENTICATION ERROR ❌] Falta 'json_keyfile_colab' en config para entorno COLAB.")
        credentials = service_account.Credentials.from_service_account_file(json_path)
    else:
        # Asumimos que para GCP (COLAB_ENTERPRISE o si se pasa un project_id distinto) se usa Secret Manager.
        secret_id = config.get("json_keyfile_GCP_secret_id")
        if not secret_id:
            raise ValueError("[AUTHENTICATION ERROR ❌] Falta 'json_keyfile_GCP_secret_id' en config para entornos GCP.")
        from google.cloud import secretmanager
        client_sm = secretmanager.SecretManagerServiceClient()
        secret_name = f"projects/{project_id}/secrets/{secret_id}/versions/latest"
        response = client_sm.access_secret_version(name=secret_name)
        secret_str = response.payload.data.decode("UTF-8")
        import json
        secret_info = json.loads(secret_str)
        credentials = service_account.Credentials.from_service_account_info(secret_info)
    return credentials
```

```python
def elevenlabs_generate_audio_str(config: dict) -> str:
    """
    Genera un archivo de audio a partir de un texto usando la API de ElevenLabs y guarda el archivo en la ruta especificada.
    Siempre se obtienen credenciales de GCP mediante la función externa `_ini_authenticate_API(config, project_id_str)` y,
    usando dichas credenciales, se accede a GCP Secret Manager para recuperar la API key de ElevenLabs.

    NOTA: Se asume que la función `_ini_authenticate_API(config_dic, project_id_str)` está definida externamente y
          retorna un objeto de credenciales GCP (service_account.Credentials).

    Args:
        config (dict):
            - text_str (str): Texto a convertir en audio (obligatorio).
            - drive_path_str (str, opcional): Ruta para guardar el archivo de audio (default: '/content/drive/MyDrive/audio_output.mp3').
            - voice_id_str (str, opcional): Identificador de la voz (default: 'default_voice_id').
            - retry_attempts (int, opcional): Número de reintentos en caso de error (default: 3).
            - retry_delay (int, opcional): Segundos a esperar entre reintentos (default: 2).
            - model_id (str, opcional): Modelo de síntesis, p. ej. 'eleven_multilingual_v2'.
            - output_format (str, opcional): Formato de salida, p. ej. 'mp3_44100_128'.

            - TTS_API_key_GCP_secret_manager_project_id (str): ID del proyecto GCP donde está el secreto de la API key.
            - TTS_API_key_GCP_secret_manager_name (str): Nombre del secreto que guarda la API key de ElevenLabs.
            - ini_environment_identificated (str, opcional): Indica el entorno (p. ej. "LOCAL", "COLAB", etc.).
            - json_keyfile_local (str, opcional): Ruta JSON de credenciales (para entornos locales).
            - json_keyfile_colab (str, opcional): Ruta JSON de credenciales (para entornos Colab).
            - json_keyfile_GCP_secret_id (str, opcional): Nombre del secret con credenciales en GCP (si aplica).

    Returns:
        str: Mensaje indicando la ruta del archivo de audio generado.

    Raises:
        ValueError: Si faltan parámetros obligatorios o se superan los reintentos.
        ElevenLabsAPIError: (Opcional) Si se dispone de este error personalizado y ocurre un error de API.
    """
    import requests
    import time
    from google.cloud import secretmanager

    # (Opcional) Intentar importar un error personalizado de ElevenLabs
    try:
        from elevenlabs.errors import ElevenLabsAPIError
    except ImportError:
        ElevenLabsAPIError = None

    # ----------------------------------------------------------------
    # 1) Validar que se haya proporcionado el texto de entrada
    # ----------------------------------------------------------------
    text_str = config.get('text_str')
    if not text_str:
        raise ValueError("[VALIDATION [ERROR ❌]] Falta 'text_str' en config.")

    # ----------------------------------------------------------------
    # 2) Validar y obtener parámetros de Secret Manager (GCP)
    # ----------------------------------------------------------------
    project_id_str = config.get("TTS_API_key_GCP_secret_manager_project_id")
    if not project_id_str:
        raise ValueError(
            "[AUTHENTICATION ERROR ❌] Falta 'TTS_API_key_GCP_secret_manager_project_id' en config."
        )

    secret_name_str = config.get("TTS_API_key_GCP_secret_manager_name")
    if not secret_name_str:
        raise ValueError(
            "[AUTHENTICATION ERROR ❌] Falta 'TTS_API_key_GCP_secret_manager_name' en config."
        )

    # ----------------------------------------------------------------
    # 3) Llamar a la función externa `_ini_authenticate_API` para
    #    obtener credenciales GCP y acceder a Secret Manager
    # ----------------------------------------------------------------
    credentials_obj = _ini_authenticate_API(config, project_id_str)

    client_sm = secretmanager.SecretManagerServiceClient(credentials=credentials_obj)
    name_str = f"projects/{project_id_str}/secrets/{secret_name_str}/versions/latest"
    response_obj = client_sm.access_secret_version(name=name_str)

    api_key_str = response_obj.payload.data.decode("UTF-8")
    if not api_key_str:
        raise ValueError("[AUTHENTICATION ERROR ❌] El secreto en GCP está vacío o no se pudo decodificar.")

    # ----------------------------------------------------------------
    # 4) Recopilar parámetros de TTS y configurar la petición
    # ----------------------------------------------------------------
    drive_path_str = config.get('drive_path_str', '/content/drive/MyDrive/audio_output.mp3')
    voice_id_str = config.get('voice_id_str', 'default_voice_id')
    retry_attempts_int = config.get('retry_attempts', 3)
    retry_delay_int = config.get('retry_delay', 2)

    model_id_str = config.get("model_id")
    output_format_str = config.get("output_format")

    print("🔹🔹🔹 [START ▶️] Generación de audio con ElevenLabs 🔹🔹🔹", flush=True)

    # Parámetros opcionales para la URL (modelo y formato)
    query_params_str = ""
    if model_id_str:
        query_params_str += f"?model_id={model_id_str}"
    if output_format_str:
        query_params_str += f"{'&' if model_id_str else '?'}output_format={output_format_str}"

    url_str = f"https://api.elevenlabs.io/v1/text-to-speech/{voice_id_str}{query_params_str}"
    headers_dic = {
        "xi-api-key": api_key_str,
        "Content-Type": "application/json"
    }
    data_dic = {
        "text": text_str
        # Agregar aquí otros campos según la documentación de ElevenLabs (p. ej. voice_settings, etc.).
    }

    # ----------------------------------------------------------------
    # 5) Llamar a la API de ElevenLabs con reintentos
    # ----------------------------------------------------------------
    response_obj = None
    attempt_int = 0
    while attempt_int < retry_attempts_int:
        try:
            response_obj = requests.post(url_str, json=data_dic, headers=headers_dic)
            if response_obj.status_code == 200:
                break
            else:
                error_msg_str = (
                    f"[API ERROR ❌] Error en la llamada a ElevenLabs: "
                    f"{response_obj.status_code} {response_obj.text}"
                )
                print(error_msg_str, flush=True)
                attempt_int += 1
                if attempt_int < retry_attempts_int:
                    print(
                        f"[RETRY 🔄] Reintentando en {retry_delay_int} s... "
                        f"(Intento {attempt_int+1}/{retry_attempts_int})",
                        flush=True
                    )
                    time.sleep(retry_delay_int)
                else:
                    if ElevenLabsAPIError:
                        raise ElevenLabsAPIError(error_msg_str)
                    else:
                        raise ValueError(error_msg_str)
        except Exception as e:
            attempt_int += 1
            print(f"[API CALL ERROR ❌] Excepción durante la llamada: {e}", flush=True)
            if attempt_int < retry_attempts_int:
                print(
                    f"[RETRY 🔄] Reintentando en {retry_delay_int} s... "
                    f"(Intento {attempt_int+1}/{retry_attempts_int})",
                    flush=True
                )
                time.sleep(retry_delay_int)
            else:
                raise ValueError(
                    f"[API CALL ERROR ❌] No se pudo generar el audio después de {retry_attempts_int} intentos: {e}"
                )

    # ----------------------------------------------------------------
    # 6) Guardar el archivo de audio en la ruta especificada
    # ----------------------------------------------------------------
    try:
        with open(drive_path_str, 'wb') as audio_file:
            audio_file.write(response_obj.content)
        print(f"[FILE SAVE SUCCESS ✅] Archivo de audio guardado en: {drive_path_str}", flush=True)
    except Exception as e:
        raise ValueError(f"[FILE SAVE ERROR ❌] No se pudo guardar el archivo: {e}")

    print("🔹🔹🔹 [FINISHED ✅] Generación de audio completada 🔹🔹🔹", flush=True)
    return f"Audio generado y guardado en: {drive_path_str}"
```

```python
import pandas as pd
from elevenlabs import ElevenLabs

def elevenlabs_voice_list(api_key):
    """
    Obtiene la lista de voces de ElevenLabs y la retorna en un DataFrame.

    Parámetros:
      - api_key: Tu clave de API de ElevenLabs.

    Retorna:
      - DataFrame con información relevante de cada voz.
    """
    # Inicializa el cliente
    client = ElevenLabs(api_key=api_key)

    # Obtiene el objeto de voces
    data = client.voices.get_all()

    # Determina si se recibe un objeto con atributo 'voices' o un diccionario
    voices = data.voices if hasattr(data, 'voices') else data.get('voices', [])

    lista_voces = []
    for voz in voices:
        # Convierte la instancia a diccionario si tiene el método 'dict'
        voz_data = voz.dict() if hasattr(voz, 'dict') else voz

        voice_info = {
            'voice_id': voz_data.get('voice_id'),
            'name': voz_data.get('name'),
            'category': voz_data.get('category'),
            'description': voz_data.get('description'),
            'preview_url': voz_data.get('preview_url'),
        }

        labels = voz_data.get('labels', {})
        voice_info['accent']  = labels.get('accent')
        voice_info['age']     = labels.get('age')
        voice_info['gender']  = labels.get('gender')
        voice_info['use_case']= labels.get('use_case')

        lista_voces.append(voice_info)

    # Retorna el DataFrame con la información recopilada
    return pd.DataFrame(lista_voces)

# Ejemplo de uso:
# api_key = "TU_API_KEY"
# df_voces = elevenlabs_voice_list(api_key)
# print(df_voces.head())
```

```python
def TTS_sintetizar_texto_str(config: dict) -> str:
    """
    Sintetiza el texto proporcionado a voz utilizando la API de Google Cloud Text-to-Speech y guarda el audio en archivos MP3.
    Permite elegir entre entrada de texto regular o SSML, según la clave 'text_input_type'.
    Para cada voz en 'voice_name_list', se genera un archivo cuyo nombre es el mismo que el de la voz.
    Se verifica que la carpeta de salida exista y se tenga permiso de escritura antes de llamar a la API.

    Args:
        config (dict):
            - text_regular (str): Texto a sintetizar en modo regular.
            - text_SSML (str): Texto en formato SSML a sintetizar.
            - text_input_type (str): Tipo de entrada, "regular" o "ssml". Default: "regular".
            - salida (str, opcional): Ruta de la carpeta de salida en Google Drive. Ejemplo: "/content/drive/MyDrive/tu_carpeta".
              Si la carpeta no existe, se creará. (Default: "output")
            - language_code (str, opcional): Código de idioma para la voz (default: "es-ES").
            - ssml_gender (str, opcional): Género de la voz cuando se usa texto regular. Opciones: "NEUTRAL", "MALE", "FEMALE" (default: "NEUTRAL").
            - voice_name_list (list, opcional): Lista de nombres específicos de voces (consulte
              https://cloud.google.com/text-to-speech/docs/voices?hl=es-419). Se genera un archivo MP3 para cada voz.
              Si se deja vacía, se utiliza la configuración por defecto.

    Returns:
        str: Mensaje indicando la ubicación de los archivos de audio generados.

    Raises:
        ValueError: Si faltan parámetros obligatorios o si 'voice_name_list' no es una lista.
        Exception: Si ocurre un error durante la verificación de escritura o la síntesis de voz.
    """
    import os

    # Subfunción: Validación y extracción de parámetros
    def _validate_config(config: dict) -> tuple:
        text_input_type = config.get("text_input_type", "regular").lower()
        if text_input_type not in ("regular", "ssml"):
            raise ValueError("[VALIDATION [ERROR ❌]] 'text_input_type' debe ser 'regular' o 'ssml'.")
        if text_input_type == "ssml":
            texto_str = config.get("text_SSML")
            if not texto_str:
                raise ValueError("[VALIDATION [ERROR ❌]] Falta 'text_SSML' en config cuando text_input_type es 'ssml'.")
        else:
            texto_str = config.get("text_regular")
            if not texto_str:
                raise ValueError("[VALIDATION [ERROR ❌]] Falta 'text_regular' en config cuando text_input_type es 'regular'.")
        salida_str = config.get("salida", "output")
        language_code_str = config.get("language_code", "es-ES")
        ssml_gender_str = config.get("ssml_gender", "NEUTRAL").upper()
        voice_name_list = config.get("voice_name_list", [])
        if not isinstance(voice_name_list, list):
            raise ValueError("[VALIDATION [ERROR ❌]] 'voice_name_list' debe ser una lista.")
        return texto_str, text_input_type, salida_str, language_code_str, ssml_gender_str, voice_name_list

    # Subfunción: Mapeo de género de voz a la enumeración de la API
    def _map_ssml_gender(ssml_gender_str: str):
        from google.cloud import texttospeech
        gender_mapping = {
            "NEUTRAL": texttospeech.SsmlVoiceGender.NEUTRAL,
            "MALE": texttospeech.SsmlVoiceGender.MALE,
            "FEMALE": texttospeech.SsmlVoiceGender.FEMALE,
        }
        return gender_mapping.get(ssml_gender_str, texttospeech.SsmlVoiceGender.NEUTRAL)

    # Subfunción: Verifica el permiso de escritura en la carpeta de salida
    def _check_write_access(folder_path: str) -> None:
        test_file = os.path.join(folder_path, "temp_write_test.txt")
        try:
            with open(test_file, "w") as f:
                f.write("test")
            os.remove(test_file)
            print(f"[FILE INFO ℹ️] Permiso de escritura confirmado en: '{folder_path}'", flush=True)
        except Exception as e:
            raise Exception(f"[FILE ERROR ❌] No se puede escribir en la carpeta '{folder_path}': {e}")

    # Subfunción: Realiza la síntesis de voz para un nombre de voz específico
    def _synthesize_speech(texto_str: str, language_code_str: str, ssml_gender_enum, voice_name_str: str, text_input_type: str) -> any:
        from google.cloud import texttospeech
        # Seleccionar la entrada de texto según el tipo: regular o SSML
        if text_input_type == "ssml":
            input_text = texttospeech.SynthesisInput(ssml=texto_str)
        else:
            input_text = texttospeech.SynthesisInput(text=texto_str)
        # Si se especifica voice_name_str, se utiliza; de lo contrario, se usa la configuración por defecto
        if voice_name_str:
            voice = texttospeech.VoiceSelectionParams(
                language_code=language_code_str,
                name=voice_name_str
            )
        else:
            voice = texttospeech.VoiceSelectionParams(
                language_code=language_code_str,
                ssml_gender=ssml_gender_enum
            )
        audio_config = texttospeech.AudioConfig(
            audio_encoding=texttospeech.AudioEncoding.MP3
        )
        print(f"[TTS INFO ℹ️] Solicitando síntesis para voz: '{voice_name_str or 'default'}'", flush=True)
        client = texttospeech.TextToSpeechClient()
        return client.synthesize_speech(
            input=input_text,
            voice=voice,
            audio_config=audio_config
        )

    # Subfunción: Guarda el contenido de audio en un archivo
    def _save_audio_file(audio_content: bytes, file_path: str) -> None:
        with open(file_path, "wb") as out:
            out.write(audio_content)
        print(f"[TTS SUCCESS ✅] Audio guardado en: '{file_path}'", flush=True)

    try:
        texto_str, text_input_type, salida_folder_str, language_code_str, ssml_gender_str, voice_name_list = _validate_config(config)
        ssml_gender_enum = _map_ssml_gender(ssml_gender_str)

        # Asegurar que la carpeta de salida exista y sea un directorio
        if not os.path.exists(salida_folder_str):
            os.makedirs(salida_folder_str, exist_ok=True)
            print(f"[FILE INFO ℹ️] Carpeta creada: '{salida_folder_str}'", flush=True)
        elif not os.path.isdir(salida_folder_str):
            raise Exception(f"[FILE ERROR ❌] La ruta '{salida_folder_str}' no es una carpeta.")

        # Verificar permiso de escritura en la carpeta de salida
        _check_write_access(salida_folder_str)

        print("🔹🔹🔹 [START ▶️] Iniciando síntesis de voz para cada voz en 'voice_name_list' 🔹🔹🔹", flush=True)
        output_files_list = []
        if not voice_name_list:
            voice_name_list = [""]

        for voice_name in voice_name_list:
            try:
                response = _synthesize_speech(texto_str, language_code_str, ssml_gender_enum, voice_name, text_input_type)
                file_name = f"{voice_name}.mp3" if voice_name else "output.mp3"
                file_path = os.path.join(salida_folder_str, file_name)
                _save_audio_file(response.audio_content, file_path)
                output_files_list.append(file_path)
            except Exception as e_voice:
                print(f"[TTS ERROR ❌] Error al sintetizar con voz '{voice_name}': {e_voice}", flush=True)
                continue

        if not output_files_list:
            raise Exception("No se generó ningún archivo de audio.")

        files_str = "\n".join(output_files_list)
        return f"Audio generado correctamente en los siguientes archivos:\n{files_str}"

    except Exception as e:
        print(f"[TTS ERROR ❌] Error durante la síntesis de voz: {e}", flush=True)
        raise e

# Ejemplo de plantilla de configuración para probar SSML (usando el ejemplo de los apuntes)
config_template = {
    "text_input_type": "ssml",  # Usar "ssml" para entrada SSML o "regular" para texto simple
    "text_SSML": (
        "<speak>\n"
        "  <p>\n"
        "    <s>Bienvenidos a la demostración de SSML.</s>\n"
        "    <s>Hoy aprenderemos a controlar la entonación y el ritmo del discurso.</s>\n"
        "  </p>\n"
        "  <p>\n"
        "    <s>\n"
        "      <prosody rate=\"medium\" pitch=\"+1st\" volume=\"medium\">\n"
        "        Este párrafo se pronuncia con una entonación ligeramente elevada.\n"
        "      </prosody>\n"
        "    </s>\n"
        "  </p>\n"
        "  <break time=\"700ms\"/>\n"
        "  <p>\n"
        "    <s>\n"
        "      <emphasis level=\"strong\">¡Atención!</emphasis> Presta mucha atención a este ejemplo.\n"
        "    </s>\n"
        "  </p>\n"
        "  <p>\n"
        "    <s>\n"
        "      La fecha de hoy es <say-as interpret-as=\"date\" format=\"dm\">25/12</say-as>.\n"
        "    </s>\n"
        "  </p>\n"
        "</speak>"
    ),
    "text_regular": "Este es un ejemplo de texto regular para síntesis de voz.",
    "salida": "/content/drive/MyDrive/test_tts3",  # Carpeta de salida en Google Drive
    "language_code": "es-ES",
    "ssml_gender": "NEUTRAL",
    "voice_name_list": voice_name_list
}

# Llamada a la función (se recomienda probar primero con pocas voces para validar la configuración)
resultado = TTS_sintetizar_texto_str(config_template)
print(resultado)
```
